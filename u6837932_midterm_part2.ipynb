```python
#algorithm program
def coin_finder(Money):
    # List of available in descending order so that we can find the largest amount first
    Money_type = [1000, 500, 100, 50, 20, 10, 5, 1]
    result = {}
    
    for coin in Money_type:
        count = Money // coin  # Determine how many coins to use to make up the amount
        if count > 0:   #if is still greater than 0 so it means that there is still some amount left
            result[coin] = count   # Store the count of each coin used
            Money -= coin * count  # Reduce the remaining amount
    
    return result

print(coin_finder(4998))
print(coin_finder(9700))

print(coin_finder(int(input("Enter the amount of money: "))))
```

```python
#Recursion program
def coin_finder_recursion(Money):
    Money_type = [1000, 500, 100, 50, 20, 10, 5, 1]   # List of available in descending order
    result = {}
    
    # Create a helper function to use recursion
    def helper(Money, index):
        if Money == 0 or index >= len(Money_type):   #Base case: if no money left or all coin types checked,
            return
        coin = Money_type[index]    # Get the current coin type
        count = Money // coin     # Determine how many coins to use to make up the amount
        if count > 0:        #if is still greater than 0 so it means that there is still some amount left
            result[coin] = count   # Store the count of each coin used
            Money -= coin * count    # Reduce the remaining amount
        helper(Money, index + 1)     # Call helper again for the next coin type
    
    helper(Money, 0)
    return result

print(coin_finder(4998))
print(coin_finder(9700))

print(coin_finder(int(input("Enter the amount of money: "))))
```
