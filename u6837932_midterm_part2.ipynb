```python
def coin_finder(N):
    # List of available in descending order so that we can find the largest amount first
    Money_type = [1000, 500, 100, 50, 20, 10, 5, 1]
    result = {}
    
    for coin in Money_type:
        count = N // coin  # Determine how many coins to use to make up the amount
        if count > 0:   #if is still greater than 0 so it means that there is still some amount left
            result[coin] = count   # Store the count of each coin used
            N -= coin * count  # Reduce the remaining amount
    
    return result

print(coin_finder(4998))
print(coin_finder(9700))

print(coin_finder(int(input("Enter the amount of money: "))))
```
