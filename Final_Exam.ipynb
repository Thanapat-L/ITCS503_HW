Question 1:<br>
Given a set of 5 numbers between 1 to 10 and a number N where N is between 1 and 100.<br>
Write a program that use all these 5 numbers to create the number N using only the<br>
following operands: +, -, x, /, and raise to the power. For example,<br>
Set_of_5 = {8, 7, 2, 3, 6}<br>
N = 37<br>
Output: The solution: 37 = (7 x (2+3)) + (8 ‚Äì 6)<br>
Hint: recursive back tracking method with brute force could be used along<br>

```python
def safe_power(a, b):
    if abs(a) > 10 or abs(b) > 10:
        return None
    try:
        val = a ** b
        if abs(val) > 1e9:
            return None
        return val
    except:
        return None


def generate_results(a, b):
    results = []
    av, ae = a
    bv, be = b

    # Addition
    results.append((av + bv, f"({ae} + {be})"))

    # Subtraction
    results.append((av - bv, f"({ae} - {be})"))
    results.append((bv - av, f"({be} - {ae})"))

    # Multiplication
    results.append((av * bv, f"({ae} * {be})"))

    # Division
    if bv != 0:
        results.append((av / bv, f"({ae} / {be})"))
    if av != 0:
        results.append((bv / av, f"({be} / {ae})"))

    # Exponent
    p1 = safe_power(av, bv)
    if p1 is not None:
        results.append((p1, f"({ae} ** {be})"))

    p2 = safe_power(bv, av)
    if p2 is not None:
        results.append((p2, f"({be} ** {ae})"))

    return results



def solve_recursive(list_num, target):
    if len(list_num) == 1:
        val, expr = list_num[0]
        if abs(val - target) < 1e-9:
            return expr
        return None

    for i in range(len(list_num)):
        for j in range(i + 1, len(list_num)):
            a = list_num[i]
            b = list_num[j]

            # Generate possible new values
            for res in generate_results(a, b):
                new_list = []

                for k in range(len(list_num)):
                    if k != i and k != j:
                        new_list.append(list_num[k])

                new_list.append(res)

                ans = solve_recursive(new_list, target)
                if ans is not None:
                    return ans

    return None



def solve_numbers(nums, target):
    arr = [(v, str(v)) for v in nums]
    ans = solve_recursive(arr, target)
    if ans is None:
        print("No solution found.")
    else:
        print("Solution found:")
        print(f"{target} = {ans}")


print("Example 1:")
solve_numbers([8, 7, 2, 3, 6], 37)

print("\nExample 2:")
solve_numbers([1, 3, 4, 6, 8], 50)

print("\nExample 3:")
solve_numbers([2, 5, 7, 9, 10], 100)
```

Question 2:<br>
Given the following fundamental cipher models for encryption:<br>
1. Shift cipher: C = (P + K) mod N<br>
2. Multiplication cipher: C = (K x P) mod N<br>
3. Polyalphabetic substitution cipher (Vigen√®re Cipher)<br>
4. Simple Columnar Permutation Transposition<br>
where<br>
‚Ä¢ C is ciphertext (number),<br>
‚Ä¢ P is plaintext (number) and<br>
‚Ä¢ K is keytext (number) and<br>
‚Ä¢ N is the number of characters in the plaintext or a block of characters<br>
The conversion of each character in the text to number is based on the ASCII table (e.g., A<br>
= 65, B = 66, a = 97, b = 98). See https://www.ascii-code.com/.<br>
Write a program that can use all 4 fundamental cipher to encrypt a plaintext to a cipher text<br>
and another program that can decrypt the cipher text to the original plaintext providing that<br>
the encrypter and decrypter know the same value of K (symmetric cryptography).<br>

```python
def encrypt_shift(plain_text, key):
    cipher_text = ""
    for character in plain_text:
        ascii_value = ord(character)
        shifted_value = (ascii_value + key) % 256
        cipher_text += chr(shifted_value)
    return cipher_text


def decrypt_shift(cipher_text, key):
    plain_text = ""
    for character in cipher_text:
        ascii_value = ord(character)
        original_value = (ascii_value - key) % 256
        plain_text += chr(original_value)
    return plain_text



def find_mod_inverse(key, modulo=256):
    for number in range(modulo):
        if (key * number) % modulo == 1:
            return number
    return None


def encrypt_multiplication(plain_text, key):
    cipher_text = ""
    for character in plain_text:
        ascii_value = ord(character)
        multiplied_value = (ascii_value * key) % 256
        cipher_text += chr(multiplied_value)
    return cipher_text


def decrypt_multiplication(cipher_text, key):
    inverse_key = find_mod_inverse(key)
    if inverse_key is None:
        raise ValueError("Key does not have a modular inverse.")

    plain_text = ""
    for character in cipher_text:
        ascii_value = ord(character)
        original_value = (ascii_value * inverse_key) % 256
        plain_text += chr(original_value)
    return plain_text



def encrypt_vigenere(plain_text, key_text):
    cipher_text = ""
    repeated_key = (key_text * ((len(plain_text) // len(key_text)) + 1))[:len(plain_text)]

    for plain_char, key_char in zip(plain_text, repeated_key):
        plain_ascii = ord(plain_char)
        key_ascii = ord(key_char)
        encrypted_value = (plain_ascii + key_ascii) % 256
        cipher_text += chr(encrypted_value)

    return cipher_text


def decrypt_vigenere(cipher_text, key_text):
    plain_text = ""
    repeated_key = (key_text * ((len(cipher_text) // len(key_text)) + 1))[:len(cipher_text)]

    for cipher_char, key_char in zip(cipher_text, repeated_key):
        cipher_ascii = ord(cipher_char)
        key_ascii = ord(key_char)
        decrypted_value = (cipher_ascii - key_ascii) % 256
        plain_text += chr(decrypted_value)

    return plain_text



def encrypt_columnar(plain_text, key_order):
    number_of_columns = len(key_order)
    number_of_rows = (len(plain_text) + number_of_columns - 1) // number_of_columns

    padded_text = plain_text + "_" * (number_of_rows * number_of_columns - len(plain_text))
    sorted_column_index = sorted(range(number_of_columns), key=lambda i: key_order[i])

    matrix = [padded_text[i*number_of_columns:(i+1)*number_of_columns] for i in range(number_of_rows)]

    cipher_text = ""
    for column in sorted_column_index:
        for row in range(number_of_rows):
            cipher_text += matrix[row][column]

    return cipher_text


def decrypt_columnar(cipher_text, key_order):
    number_of_columns = len(key_order)
    number_of_rows = (len(cipher_text) + number_of_columns - 1) // number_of_columns
    sorted_column_index = sorted(range(number_of_columns), key=lambda i: key_order[i])

    matrix = [[""] * number_of_columns for _ in range(number_of_rows)]

    current_index = 0
    for column in sorted_column_index:
        for row in range(number_of_rows):
            matrix[row][column] = cipher_text[current_index]
            current_index += 1

    plain_text = ""
    for row in range(number_of_rows):
        plain_text += "".join(matrix[row])

    return plain_text.replace("_", "")



text1 = "SUP MATE"
key1 = 5
cipher1 = encrypt_shift(text1, key1)
plain1 = decrypt_shift(cipher1, key1)
print("Plaintext :", text1)
print("Encrypted :", cipher1)
print("Decrypted :", plain1)

print("\nMultiplication Cipher")
text2 = "SECURITY"
key2 = 7
cipher2 = encrypt_multiplication(text2, key2)
plain2 = decrypt_multiplication(cipher2, key2)
print("Plaintext :", text2)
print("Encrypted :", cipher2)
print("Decrypted :", plain2)

print("\nVigenere Cipher")
text3 = "okay sure"
key3 = "KEY"
cipher3 = encrypt_vigenere(text3, key3)
plain3 = decrypt_vigenere(cipher3, key3)
print("Plaintext :", text3)
print("Encrypted :", cipher3)
print("Decrypted :", plain3)

print("\nColumnar Transposition")
text4 = "OKAY GOOGLE"
key4 = [3, 1, 4, 2]
cipher4 = encrypt_columnar(text4, key4)
plain4 = decrypt_columnar(cipher4, key4)
print("Plaintext :", text4)
print("Encrypted :", cipher4)
print("Decrypted :", plain4)
```

```python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

def mod_inverse(e, phi):
    for d in range(1, phi):
        if (d * e) % phi == 1:
            return d
    return None

def generate_keys(p, q):
    n = p * q
    phi_n = (p - 1) * (q - 1)

    for e in range(2, phi_n):
        if gcd(e, phi_n) == 1:
            break

    d = mod_inverse(e, phi_n)
    return (e, n), (d, n)



def encrypt_rsa(plain_number, public_key):
    e, n = public_key
    return pow(plain_number, e, n)

def decrypt_rsa(cipher_number, private_key):
    d, n = private_key
    return pow(cipher_number, d, n)



def sign_message(hash_value, private_key):
    d, n = private_key
    return pow(hash_value, d, n)

def verify_signature(signature_value, public_key):
    e, n = public_key
    return pow(signature_value, e, n)



print("SAMPLE 1: p=61, q=53")

public_key, private_key = generate_keys(61, 53)

message = 42
cipher = encrypt_rsa(message, public_key)
dec = decrypt_rsa(cipher, private_key)

print("Public :", public_key)
print("Private:", private_key)
print("Plaintext:", message)
print("Ciphertext:", cipher)
print("Decrypted:", dec)



print("\nSAMPLE 2: p=47, q=71")

public_key2, private_key2 = generate_keys(47, 71)

message2 = 99
cipher2 = encrypt_rsa(message2, public_key2)
dec2 = decrypt_rsa(cipher2, private_key2)

print("Public :", public_key2)
print("Private:", private_key2)
print("Plaintext:", message2)
print("Ciphertext:", cipher2)
print("Decrypted:", dec2)



print("\nSAMPLE 3: Digital Signature with p=59, q=83")

alice_public, alice_private = generate_keys(59, 83)

msg = "Hi my name is Justin Azanzmanhu"
msg_hash = sum(ord(c) for c in msg)

signature = sign_message(msg_hash, alice_private)
verified = verify_signature(signature, alice_public)

print("Message:", msg)
print("Hash   :", msg_hash)
print("Signature:", signature)
print("Verified Hash:", verified)

if verified == msg_hash:
    print("Signature is VALID (Signed by Alice)")
else:
    print("Signature is INVALID")
```

Question 4: <br>
Mis√®re Tic-Tac-Toe (Avoid Losing Tic-Tac-Toe) is a variant of normal Tic-Tac-Toe).   <br>
Rules: <br>
‚Ä¢ 3√ó3 grid, players alternate placing X and O as usual. <br>
‚Ä¢ Goal is inverted: the player who first makes 3 in a row LOSES. <br>
‚Ä¢ If the board fills with no 3-in-a-row, it‚Äôs a draw. <br>
Write an interactive AI program that can play against a player using the minimax function to <br>
be in the winning position. <br>

```python
def create_board():
    return [" "] * 9

def print_board(b):
    print("Current board:")
    print(" " + b[0] + " | " + b[1] + " | " + b[2])
    print("---+---+---")
    print(" " + b[3] + " | " + b[4] + " | " + b[5])
    print("---+---+---")
    print(" " + b[6] + " | " + b[7] + " | " + b[8])
    print()

def three_in_a_row(b, s):
    lines = [(0,1,2),(3,4,5),(6,7,8),
             (0,3,6),(1,4,7),(2,5,8),
             (0,4,8),(2,4,6)]
    for a, b1, c in lines:
        if b[a] == s and b[b1] == s and b[c] == s:
            return True
    return False

def board_full(b):
    return all(x != " " for x in b)

def evaluate(b, bot, human):
    if three_in_a_row(b, human):  
        return +1
    if three_in_a_row(b, bot):    
        return -1
    if board_full(b):
        return 0
    return None

def minimax(b, bot_turn, bot, human):
    score = evaluate(b, bot, human)
    if score is not None:
        return score

    if bot_turn:
        best = -999
        for i in range(9):
            if b[i] == " ":
                b[i] = bot
                best = max(best, minimax(b, False, bot, human))
                b[i] = " "
        return best
    else:
        best = 999
        for i in range(9):
            if b[i] == " ":
                b[i] = human
                best = min(best, minimax(b, True, bot, human))
                b[i] = " "
        return best

def bot_move(b, bot, human):
    best = -999
    move = None
    for i in range(9):
        if b[i] == " ":
            b[i] = bot
            val = minimax(b, False, bot, human)
            b[i] = " "
            if val > best:
                best = val
                move = i
    return move

def human_move(b, s):
    while True:
        pos = input("Enter move (1-9): ")
        if not pos.isdigit():
            print("Numbers only.")
            continue
        pos = int(pos) - 1
        if pos < 0 or pos > 8:
            print("Choose 1‚Äì9.")
        elif b[pos] != " ":
            print("That spot is taken.")
        else:
            b[pos] = s
            break

def play():
    b = create_board()
    bot = "O"
    human = "X"
    turn = "human"

    print("Mis√®re Tic-Tac-Toe (avoid making 3 in a row!)")
    print_board(b)

    while True:
        if turn == "You":
            human_move(b, human)
            print_board(b)
            r = evaluate(b, bot, human)
            if r is not None:
                if r == +1: print("You made 3 in a row. You LOSE.")
                elif r == -1: print("Bot made 3 in a row. You WIN!")
                else: print("Draw.")
                break
            turn = "bot"

        else:
            print("Bot is thinking...")
            m = bot_move(b, bot, human)
            if m is None:
                print("Draw.")
                break
            b[m] = bot
            print_board(b)
            r = evaluate(b, bot, human)
            if r is not None:
                if r == +1: print("You made 3 in a row. You LOSE.")
                elif r == -1: print("Bot made 3 in a row. You WIN!")
                else: print("Draw.")
                break
            turn = "You"

play()
```
